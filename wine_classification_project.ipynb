```python
# Libraries
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
import joblib
%matplotlib inline

def show_sample(data):
    print('                                             Initial Visualization                                                ')
    print('------------------------------------------------------------------------------------------------------------------')
    print('                                               Sample of 10 rows                                                  ')
    display(data.sample(10))
    print('==================================================================================================================')
    return None

def show_data_info(data):
    print('                                                   Data info                                                      ')
    print(data.info())
    print('==================================================================================================================')
    return None

def county_empty_columns(data):
    print('                                            Number of null columns                                                ')
    print(data.isnull().sum())
    print('==================================================================================================================')
    return None


def features_corr(data):
    # Pearson Correlation 
    print('                                              Pearson Correlation                                                 ')
    plt.figure(figsize=(15,10))
    cor=data.corr()
    sns.heatmap(cor, annot=True, cmap='crest')
    plt.show()

    # Most relevant features
    cor_target = abs(cor['WineVariety'])
    relevant_features = cor_target
    
    relevant_features_60 = cor_target[cor_target>0.60]
    print('==================================================================================================================')
    print('                               The 5 most relevant features (correlation > 0.6)                                   ')
    print('\n')
    print(relevant_features_60)
    print('==================================================================================================================')
    return None

def plot_WineVariety_statistics(data):
    print('                                                Wine Variety                                                      ')
    # Get the label column
    label = data['WineVariety']

    # Get statistics
    label_min = label.min()
    label_max = label.max()
    label_mean = label.mean()
    label_median = label.median()
    label_mode = label.mode()[0]

    print('Minimum:{:.2f}\nMean:{:.2f}\nMedian:{:.2f}\nMode:{:.2f}\nMaximum:{:.2f}\n'.format(label_min, 
                                                                                                 label_max, 
                                                                                                 label_mean, 
                                                                                                 label_median, 
                                                                                                 label_mode))
    # Create subplots
    fig, ax = plt.subplots(2,1, figsize=(7,9))

    # Plot histogram
    ax[0].hist(label)

    # Add lines for the min, max, mean, median, and mode
    ax[0].axvline(label_min, color='gray', linestyle='dashed', linewidth=2)
    ax[0].axvline(label_max, color='gray', linestyle='dashed', linewidth=2)
    ax[0].axvline(label_mean, color='cyan', linestyle='dashed', linewidth=2)
    ax[0].axvline(label_median, color='magenta', linestyle='dashed', linewidth=2)
    ax[0].axvline(label_mode, color='yellow', linestyle='dashed', linewidth=2)

    # Plot boxplot
    ax[1].boxplot(label, vert=False)

    # Titles
    for i in range(0,2):
        ax[i].set_title('Data Distribution')
        ax[i].set_xlabel('Price')
        ax[i].set_ylabel('Frequency')

    # Show
    plt.show()
    print('==================================================================================================================')
    return None

def plot_features_statistics(data):
    features = ['Alcohol','Malic_acid','Ash','Alcalinity','Magnesium','Phenols','Flavanoids','Nonflavanoids','Proanthocyanins','Color_intensity','Hue','OD280_315_of_diluted_wines','Proline']
    label = 'WineVariety'

    print('                                              Numeric Features                                                    ')
    for col in features:
        data.boxplot(column=col, by=label, figsize=(6,6))
        plt.title(col)
    plt.show()
    print('==================================================================================================================')
    return None

def machine_learning(data):
    print('                                              Machine Learning                                                    ')
    # Input features
    X = data.drop('WineVariety', axis=1)

    # Target 
    y = data['WineVariety']

    # Split data 70%-30% into training set and test set
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
    print('Training set: %d rows \nTest set: %d rows' % (X_train.shape[0], X_test.shape[0]))

    # Set regularization rate
    reg = 0.1

    # Fit a Logistic Regression model on the training set
    lrm = LogisticRegression(C=1/reg, solver='lbfgs', multi_class='auto', max_iter=10000).fit(X_train, y_train)

    # Predict
    y_pred = lrm.predict(X_test)

    # Evaluate the model
    print('------------------------------------------------------------------------------------------------------------------')
    print('                                              Evaluate metrics                                                    ')
    print('Accuracy: ', accuracy_score(y_test, y_pred))
    print('Overall Precision: ', precision_score(y_test, y_pred, average='macro'))
    print('Overall Recall: ', recall_score(y_test, y_pred, average='macro'))
    
    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    classes = ['Variety A','Variety B','Variety C']
    plt.imshow(cm, interpolation="nearest", cmap=plt.cm.Blues)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    plt.title('Confusion Matrix')
    plt.xlabel("Predicted Variety")
    plt.ylabel("Actual Variety")
    plt.show()
    
    # Roc curve
    # Get predictions from test data
    y_scores = lrm.predict_proba(X_test)

    # Get ROC metrics for each class
    fpr = {}
    tpr = {}
    thresh ={}
    for i in range(len(classes)):    
        fpr[i], tpr[i], thresh[i] = roc_curve(y_test, y_scores[:,i], pos_label=i)

    # Plot the ROC chart
    plt.plot(fpr[0], tpr[0], linestyle='--',color='orange', label=classes[0] + ' vs Rest')
    plt.plot(fpr[1], tpr[1], linestyle='--',color='green', label=classes[1] + ' vs Rest')
    plt.plot(fpr[2], tpr[2], linestyle='--',color='blue', label=classes[2] + ' vs Rest')
    plt.title('Multiclass ROC curve')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive rate')
    plt.legend(loc='best')
    plt.show()
    
    # Roc auc score
    y_scores = lrm.predict_proba(X_test)

    auc = roc_auc_score(y_test, y_scores, multi_class='ovr')
    print('Average AUC:', auc)
    print('==================================================================================================================')

    print('                             New data to test the Artificial Intelligence (AI)                                    ')
    # New data to test the Artificial Intelligence (AI)
    d = {'Alcohol': [13.72, 12.37],
    'Malic_acid': [1.43, 0.94],
    'Ash': [2.5, 1.36],
    'Alcalinity': [16.7, 10.6],
    'Magnesium': [108, 88],
    'Phenols': [3.4, 1.98],
    'Flavanoids': [3.67, 0.57],
    'Nonflavanoids': [0.19, 0.28],
    'Proanthocyanins': [2.04, 0.42],
    'Color_intensity': [6.8, 1.95],
    'Hue': [0.89, 1.05],
    'OD280_315_of_diluted_wines': [2.87, 1.82],
    'Proline': [1285, 520]}

    # Transform the array 
    new_data = pd.DataFrame(d)
    display(new_data)

    X = new_data

    # Save the Logistic Regression model as a pickle file
    filename = './wine-classification-project.pkl'
    joblib.dump(lrm, filename)

    # Load the model
    loaded_model = joblib.load(filename)

    # Predict the column 'WineVariety' of the new_data
    results = loaded_model.predict(X)

    # Transform the results array in a DataFrame
    results = pd.DataFrame(results)

    # Change column name
    results.columns = ['WineVariety']
    
    print('                                    Results for the two new wine samples                                          ')
    # Show results
    display(results)

    return None

def data_collect (path):
    # Load dataset
    data = pd.read_csv(path)
    
    show_sample(data)
    
    show_data_info(data)
    
    county_empty_columns(data)
    
    return data


def data_transform (data):
    
    features_corr(data)
    
    plot_WineVariety_statistics(data)
    
    plot_features_statistics(data)
    
    machine_learning(data)
    
    return data


if __name__ == '__main__':
    #ETL
    data_raw = data_collect ('wine.csv')
    
    data_processing = data_transform (data_raw)
```
